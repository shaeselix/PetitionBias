```python
import re
import collections
import random
import math
```

```python
def allsummarypages(summaryio):
    p1 = re.compile('(?<= \(|; )[^;()]*: p\. [^();]*(?=\)|;)')
    p2 = re.compile(': p\. ')
    p3 = re.compile('l\.[^p$]*')
    p4 = re.compile('[0-9-]+')
    p5 = re.compile('-')
    pcdict = {} 
    for c in p1.finditer(summaryio.read()):
        citesplit = p2.split(c.group(0))
        if citesplit[0] not in pcdict:
            pcdict[citesplit[0]] = collections.Counter()
        for s in p4.finditer(p3.sub('',citesplit[1])):
            if '-' in s.group(0):
                r = p5.split(s.group(0))
                for p in range(int(r[0]), int(r[1]) + 1):
                    pcdict[citesplit[0]][p] += 1
            else:
                pcdict[citesplit[0]][int(s.group(0))] += 1    
    summaryio.close()
    return(pcdict)
```

```python
testingset = random.sample(list(pages.keys()),0)
trainingset = [pg for pg in pages.keys() if pg not in testingset]
```

```python
N_t = len(trainingset)
N_r = sum(1 for pg in trainingset if pg in relcount2.keys())
N_i = N_t - N_r
prior_r = float(N_r/N_t)
prior_i = float(N_i/N_t)
print('Total Training Pages: %s' % N_t)
print('Relevant Training Pages: %s, Prior: %s' % (N_r,prior_r))
print('Irrelevant Training Pages: %s, Prior: %s' % (N_i,prior_i))
```

```python
relcounter = collections.Counter()
irrelcounter = collections.Counter()
p6 = re.compile('[a-z]+')
for pg in trainingset:
    if pg in relcount2.keys():
        for w in p6.finditer(pages[pg].lower()):
            relcounter[w.group(0)] += 1
    else:
        for w in p6.finditer(pages[pg].lower()):
            irrelcounter[w.group(0)] += 1
```

```python
vocabulary = (relcounter + irrelcounter).keys()
```

```python
#vocabulary
```

```python
tw_r = sum(relcounter.values())
tw_i = sum(irrelcounter.values())
v_sum = len(vocabulary)
print('total word counts r: %s\ntotal word counts i: %s\ntotal vocab: %s' % (tw_r,tw_i,v_sum))
```

```python
condprob = {}
wordlist = []
for word in vocabulary:
    rcp = (relcounter[word] + 1)/(tw_r + v_sum)
    icp = (irrelcounter[word] + 1)/(tw_i + v_sum)
    condprob[word] = (rcp, icp)
    wordlist.append((rcp/icp,word))
```

```python
predict = {}
correct = 0
for pg in testingset:
    score_r = math.log(prior_r)
    score_i = math.log(prior_i)
    for w in p6.finditer(pages[pg].lower()):
        word = w.group(0)
        print(word)
        if word in vocabulary:
            print('in')
            print('(%s , %s)' % (math.log(condprob[word][0]),math.log(condprob[word][1])))
            score_r += math.log(condprob[word][0])
            score_i += math.log(condprob[word][1])    
    predd = {}
    predd['prediction'] = 'I'
    if score_r > score_i:
        predd['prediction'] = 'R'
    predd['actual'] = 'I'
    if pg in relcount.keys():
        predd['actual'] = 'R'
    predd['correct'] = True
    if predd['prediction'] != predd['actual']:
        predd['correct'] = False
    else:
        correct += 1
    predict[pg] = predd
predict['score'] = float(correct/len(testingset))
```
